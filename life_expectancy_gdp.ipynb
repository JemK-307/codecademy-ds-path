```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read file
file = "all_data.csv"
```

```python
# read file csv into DataFrame
# print out the first 5 rows
df = pd.read_csv(file)
df.head()
```

```python
# let's take a first look at this data
df.info()
```

This data has 4 columns with 96 rows; No null values in this data

```python
# Which country presented in this data

df['Country'].unique()
```

```python
# Which year presented in this data

df['Year'].unique()
```

So there're 6 countries in this data and the year represented in this data is range from 2000 to 2015

```python
# From 4 cols; Change the name of the column to be easy to use

df.rename(columns={'Life expectancy at birth (years)':'LifeExp'}, inplace=True)

# and change the country name United State to USA
df['Country'].replace({'United States of America':'USA'}, inplace=True)
```

```python
# by average which country has the highest GDP

sns.barplot(y='Country', x='GDP', data=df, palette="magma");
```

```python
# by average which country has the highest Life expectancy

sns.barplot(y='Country', x='LifeExp', data=df, palette="magma");
```

From the bar chart...we can see that there's a huge gap in the average GDP by country. But for the average of Life expectancy seems similar; except for Zimbabwe.

```python
# compare GDP and LifeExp by Year

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.lineplot(x='Year', y='GDP', hue='Country', data=df);
plt.title('GDP per Country in each year')
plt.xlabel('GDP in trillions')
plt.legend(loc='upper left')

plt.subplot(1,2,2)
sns.lineplot(x='Year', y='LifeExp', hue='Country', data=df);
plt.title('Life Exp per Country in each year')
plt.xlabel('Life expectancy(YEAR)')

plt.show()
```

```python
chartGDP = sns.FacetGrid(data=df, col='Country', col_wrap=3, hue='Country')
chartGDP.map_dataframe(sns.lineplot, x='Year', y='GDP');
chartGDP.set_axis_labels("Year", "GDP in Trillions");
```

```python
chartLife = sns.FacetGrid(data=df, col='Country', col_wrap=3, hue='Country')
chartLife.map_dataframe(sns.lineplot, x='Year', y='LifeExp');
chartLife.set_axis_labels("Year", "Life Expectancy in Years");
```

In the line chart...

For the life expectancy chart >> it shows that there's a growth since 2000 but Zimbabwehas the significant growth over years.

For the GDP chart >> USA has the highest GDP although there's a decrease around year 2009 but USA is still leading in GDP; but China has the greatest growth in last 15 years.

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.boxplot(x='Country', y='GDP', data=df);

plt.subplot(1,2,2)
sns.boxplot(x='Country', y='LifeExp', data=df);

plt.show()
```

Let's look at the boxplot...

China has the widest IQR in term of GDP when Zimbabwe has the widest IQR in term of Life Expectancy.

and you will see that even Chile has the low GDP but their life expectancy is so high...and if you compare this boxplot with line plot you will see that the distribution in life expectancy is narrow for Chile USA Germany China and Mexico..because of the trend for life expectancy for these countries are steadily increase.
